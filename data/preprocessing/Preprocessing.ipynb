```python
## Phase 3: Outlier Detection and Analysis

sample_features = [
    "eeg_feature_1",
    "eeg_feature_2",
    "eeg_feature_3",
    "eye_feature_1",
    "eye_feature_2",
    "eye_feature_3"
]
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df[sample_features]
)

plt.title("Outlier Detection using Boxplots")

plt.xticks(rotation=45)

plt.show()
```

```python
def count_outliers_iqr(series):

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = (
        (series < lower) |
        (series > upper)
    )

    return outliers.sum()
for col in sample_features:

    count = count_outliers_iqr(df[col])

    print(f"{col}: {count}")
for col in sample_features:

    count = count_outliers_iqr(df[col])

    percentage = (
        count / len(df)
    ) * 100

    print(
        f"{col}: {percentage:.2f}%"
    )
```
