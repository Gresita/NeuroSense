```python
## Phase 3: Outlier Detection and Analysis

sample_features = [
    "eeg_feature_1",
    "eeg_feature_2",
    "eeg_feature_3",
    "eye_feature_1",
    "eye_feature_2",
    "eye_feature_3"
]
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df[sample_features]
)

plt.title("Outlier Detection using Boxplots")

plt.xticks(rotation=45)

plt.show()
```

```python
def count_outliers_iqr(series):

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = (
        (series < lower) |
        (series > upper)
    )

    return outliers.sum()
for col in sample_features:

    count = count_outliers_iqr(df[col])

    print(f"{col}: {count}")
for col in sample_features:

    count = count_outliers_iqr(df[col])

    percentage = (
        count / len(df)
    ) * 100

    print(
        f"{col}: {percentage:.2f}%"
    )
```

```python
## Phase 4: Feature Engineering
eeg_features = [
    col for col in df.columns
    if col.startswith("eeg_")
]

eye_features = [
    col for col in df.columns
    if col.startswith("eye_")
]

print("EEG Features:", len(eeg_features))
print("Eye Features:", len(eye_features))

df["avg_eeg"] = df[eeg_features].mean(axis=1)
df["std_eeg"] = df[eeg_features].std(axis=1)
df["max_eeg"] = df[eeg_features].max(axis=1)
df["min_eeg"] = df[eeg_features].min(axis=1)

df["avg_eye"] = df[eye_features].mean(axis=1)
df["std_eye"] = df[eye_features].std(axis=1)
df["max_eye"] = df[eye_features].max(axis=1)
df["min_eye"] = df[eye_features].min(axis=1)

df["eeg_eye_ratio"] = (
    df["avg_eeg"] /
    (df["avg_eye"] + 1e-6)
)
```

```python
df["eeg_range"] = df["max_eeg"] - df["min_eeg"]
df["eye_range"] = df["max_eye"] - df["min_eye"]

df["eeg_stability"] = df["avg_eeg"] / (df["std_eeg"] + 1e-6)
df["eye_stability"] = df["avg_eye"] / (df["std_eye"] + 1e-6)

df["phys_activity"] = df["avg_eeg"] + df["avg_eye"]
```

```python
#Phase 6 - Feature Scaling
exclude_cols = [
    "subject",
    "session",
    "trial",
    "sample",
    "label",
    "emotion"
]
feature_cols = [
    col for col in df.columns
    if col not in exclude_cols
]

X = df[feature_cols]

print(X.shape)
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

X_scaled = scaler.fit_transform(X)
print(X_scaled.shape)
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=feature_cols
)

X_scaled_df.head()
```

```python
#PCA
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA

# kontroll
print("Missing values before imputation:", X.isnull().sum().sum())

imputer = SimpleImputer(strategy="mean")

X_imputed = imputer.fit_transform(X)

print("Missing values after imputation:", np.isnan(X_imputed).sum())

scaler = RobustScaler()

X_scaled = scaler.fit_transform(X_imputed)

print("Scaled shape:", X_scaled.shape)
```

```python
pca = PCA()

X_pca = pca.fit_transform(X_scaled)

explained_variance = np.cumsum(
    pca.explained_variance_ratio_
)

n_components = np.argmax(
    explained_variance >= 0.95
) + 1

print(f"Components needed for 95% variance: {n_components}")
```
